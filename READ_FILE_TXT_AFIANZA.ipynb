```python
from pathlib import Path
import pandas as pd

# =====================================================
# üìÖ FECHA DE REPORTE
# =====================================================
fecha_reporte = "2025-11-21"

# Convertir a YYYYMMDD sin guiones
fecha_str = fecha_reporte.replace("-", "")
print("üìÖ Fecha reporte:", fecha_str)

# =====================================================
# üìÅ CONFIGURACI√ìN DE RUTAS (USANDO PATHLIB)
# =====================================================

# Carpeta base
base_dir = (
    Path.home()
    / "EF Securitizadora"
    / "Operaciones - Documentos"
    / "Documentos"
    / "Operaciones"
    / "REDEC"
    / "GLOBAL_AFIANZA"
)

# Carpetas
ruta_txt_dir = base_dir / "archivo_txt"
ruta_xlsx_dir = base_dir / "archivo_xlsx"

# Crear carpetas si no existen
ruta_txt_dir.mkdir(parents=True, exist_ok=True)
ruta_xlsx_dir.mkdir(parents=True, exist_ok=True)

# Archivo TXT esperado seg√∫n la fecha
archivo_txt = f"RDC01_{fecha_str}.TXT"
ruta_txt = ruta_txt_dir / archivo_txt

# Archivo Excel con tu nuevo formato
archivo_excel = f"RDC01_GLOBAL_AFIANZA_{fecha_str}.xlsx"
ruta_excel = ruta_xlsx_dir / archivo_excel

print("üìÅ TXT:", ruta_txt)
print("üìÅ XLSX:", ruta_excel)

# =====================================================
# üìÑ ESTRUCTURA DEL ARCHIVO RDC01
# =====================================================

columnas = [
    "RUT", "TIPO_PERSONA", "CODIGO_OPERACION", "OPERACION_TITULO_III",
    "TIPO_DEUDOR", "TIPO_DE_CREDITO", "FECHA_OTORGAMIENTO", "CARGA_FINANCIERA",
    "FECHA_EXTINCION_OPERACION", "VALOR_GARANTIA_INMOBILIARIA",
    "VALOR_GARANTIA_MOBILIARIA", "VALOR_GARANTIA_FINANCIERA",
    "VALOR_GARANTIA_AVAL_FIANZA", "MONTO_ORIGINAL_OPERACION",
    "MONTO_ACTUAL_OPERACION", "MONTO_AL_DIA",
    "MONTO_MORA_1_TRAMO", "MONTO_MORA_2_TRAMO", "MONTO_MORA_3_TRAMO",
    "MONTO_MORA_4_TRAMO", "MONTO_MORA_5_TRAMO", "MONTO_MORA_6_TRAMO",
    "MONTO_MORA_7_TRAMO", "MONTO_MORA_8_TRAMO", "MONTO_MORA_9_TRAMO",
    "MORA_ACTUAL", "DEUDA_RENEGOCIADA", "DEUDA_ACELERADA"
]

anchos = [
    10, 1, 30, 1, 1, 2, 8, 15, 8,
    15, 15, 15, 15, 15, 15,
    15, 15, 15, 15, 15, 15,
    15, 15, 15, 15,
    4, 1, 1
]

columnas_float = [
    'CARGA_FINANCIERA', 'VALOR_GARANTIA_INMOBILIARIA', 'VALOR_GARANTIA_MOBILIARIA',
    'VALOR_GARANTIA_FINANCIERA', 'VALOR_GARANTIA_AVAL_FIANZA',
    'MONTO_ORIGINAL_OPERACION', 'MONTO_ACTUAL_OPERACION', 'MONTO_AL_DIA',
    'MONTO_MORA_1_TRAMO', 'MONTO_MORA_2_TRAMO', 'MONTO_MORA_3_TRAMO',
    'MONTO_MORA_4_TRAMO', 'MONTO_MORA_5_TRAMO', 'MONTO_MORA_6_TRAMO',
    'MONTO_MORA_7_TRAMO', 'MONTO_MORA_8_TRAMO', 'MONTO_MORA_9_TRAMO',
    'MORA_ACTUAL'
]

columnas_fecha = ['FECHA_OTORGAMIENTO', 'FECHA_EXTINCION_OPERACION']

# =====================================================
# üîç VERIFICAR TXT
# =====================================================

if not ruta_txt.exists():
    print("‚ùå No se encontr√≥ el archivo TXT para la fecha:", fecha_str)
    print("Ruta buscada:", ruta_txt)
else:
    print("‚úÖ Archivo TXT encontrado. Procesando...\n")

    # =====================================================
    # üì• LECTURA ARCHIVO ANCHO FIJO
    # =====================================================
    df = pd.read_fwf(
        ruta_txt,
        widths=anchos,
        names=columnas,
        dtype=str,
        skiprows=1
    )

    df = df.applymap(lambda x: x.strip() if isinstance(x, str) else x)

    for col in columnas_float:
        df[col] = pd.to_numeric(df[col], errors="coerce")

    for col in columnas_fecha:
        df[col] = pd.to_datetime(df[col], format="%Y%m%d", errors="coerce")

    # =====================================================
    # üíæ EXPORTAR A EXCEL
    # =====================================================
    df.to_excel(ruta_excel, index=False)

    print(f"‚úÖ Excel creado correctamente:\n{ruta_excel}")
```
