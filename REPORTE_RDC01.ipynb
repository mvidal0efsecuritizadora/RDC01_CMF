```python
# =====================================================
# ðŸ§¾ FUNCIÃ“N PARA GENERAR ARCHIVO RDC01 EN FORMATO TXT CMF
# =====================================================
def generar_RDC01_txt(df_RDC01, fecha_reporte, ruta_txt, codigo_institucion="0000000782", nombre_archivo=None):

    # --- Convertir ruta_txt a pathlib ---
    ruta_txt = Path(ruta_txt)
    ruta_txt.mkdir(parents=True, exist_ok=True)

    # --- Preparar fecha ---
    fecha_reporte = pd.to_datetime(fecha_reporte)
    fecha_str = fecha_reporte.strftime("%Y%m%d")

    # Nombre final del archivo
    if nombre_archivo is None:
        nombre_archivo = f"RDC01_{fecha_str}.txt"

    # Ruta completa de salida
    ruta_salida_txt = ruta_txt / nombre_archivo

    # ----------- Funciones auxiliares -----------
    def fmt_num(val, length):
        val = 0 if pd.isna(val) else int(val)
        return str(val).rjust(length, "0")[:length]

    def fmt_str(val, length):
        val = "" if pd.isna(val) else str(val)
        return val.ljust(length, " ")[:length]

    def fmt_date(val):
        if pd.isna(val) or val == "" or val == "NaT":
            return "00000000"
        d = pd.to_datetime(val, errors="coerce")
        return "00000000" if pd.isna(d) else d.strftime("%Y%m%d")

    # =====================================================
    # ðŸ§± CABECERA (322 caracteres)
    # =====================================================
    cabecera = (
        codigo_institucion.rjust(10, "0") +   # CÃ³digo instituciÃ³n (10)
        "RDC01".ljust(5, " ") +               # Tipo archivo (5)
        fecha_str +                           # Fecha (8)
        " " * 299                             # Filler (299)
    )
    assert len(cabecera) == 322

    lineas = [cabecera]

    # =====================================================
    # ðŸ“„ DETALLES
    # =====================================================
    for _, row in df_RDC01.iterrows():
        linea = (
            fmt_str(row["RUT"], 10) +
            fmt_num(row["TIPO_PERSONA"], 1) +
            fmt_str(row["CODIGO_OPERACION"], 30) +
            fmt_num(row["OPERACION_TITULO_III"], 1) +
            fmt_num(row["TIPO_DEUDOR"], 1) +
            fmt_num(row["TIPO_CREDITO"], 2) +
            fmt_date(row["FECHA_OTORGAMIENTO"]) +
            fmt_num(row["CARGA_FINANCIERA"], 15) +
            fmt_date(row["FECHA_EXTINCION_OPERACION"]) +
            fmt_num(row["VALOR_GARANTIA_INMOBILIARIA"], 15) +
            fmt_num(row["VALOR_GARANTIA_MOBILIARIA"], 15) +
            fmt_num(row["VALOR_GARANTIA_FINANCIERA"], 15) +
            fmt_num(row["VALOR_GARANTIA_AVAL_FIANZA"], 15) +
            fmt_num(row["MONTO_ORIGINAL_OPERACION"], 15) +
            fmt_num(row["MONTO_ACTUAL_OPERACION"], 15) +
            fmt_num(row["MONTO_AL_DIA"], 15) +
            fmt_num(row["MONTO_MORA_1_TRAMO"], 15) +
            fmt_num(row["MONTO_MORA_2_TRAMO"], 15) +
            fmt_num(row["MONTO_MORA_3_TRAMO"], 15) +
            fmt_num(row["MONTO_MORA_4_TRAMO"], 15) +
            fmt_num(row["MONTO_MORA_5_TRAMO"], 15) +
            fmt_num(row["MONTO_MORA_6_TRAMO"], 15) +
            fmt_num(row["MONTO_MORA_7_TRAMO"], 15) +
            fmt_num(row["MONTO_MORA_8_TRAMO"], 15) +
            fmt_num(row["MONTO_MORA_9_TRAMO"], 15) +
            fmt_num(row["MORA_ACTUAL"], 4) +
            fmt_num(row["DEUDA_RENEGOCIADA"], 1) +
            fmt_num(row["DEUDA_ACELERADA"], 1)
        )

        if len(linea) != 322:
            raise ValueError(f"âŒ LÃ­nea tiene {len(linea)} caracteres (deben ser 322).")

        lineas.append(linea)

    # =====================================================
    # ðŸ’¾ GUARDAR ARCHIVO TXT
    # =====================================================
    with open(ruta_salida_txt, "w", encoding="utf-8") as f:
        f.write("\n".join(lineas))

    print("===========================================")
    print(f"âœ… TXT generado correctamente")
    print(f"ðŸ“ Ruta: {ruta_salida_txt}")
    print(f"ðŸ“¦ Registros totales (incluye cabecera): {len(lineas)}")
    print("===========================================")

    return ruta_salida_txt
```

```python
ruta_txt = (
    Path.home()
    / "EF Securitizadora"
    / "Operaciones - Documentos"
    / "Documentos"
    / "Operaciones"
    / "REDEC"
    / "REPORTE_RDC01"
    / "archivo_txt"
)

generar_RDC01_txt(
    df_reporte_RDC01,
    fecha_reporte="2025-11-21",
    ruta_txt=ruta_txt,
    codigo_institucion="0000000782"
)
```

```python
def generar_caratula_rdc01(df, institucion, codigo, fecha_ddmmaaaa, ruta_guardado):
    campos_mora = [
        "MONTO_MORA_1_TRAMO", "MONTO_MORA_2_TRAMO", "MONTO_MORA_3_TRAMO",
        "MONTO_MORA_4_TRAMO", "MONTO_MORA_5_TRAMO", "MONTO_MORA_6_TRAMO",
        "MONTO_MORA_7_TRAMO", "MONTO_MORA_8_TRAMO", "MONTO_MORA_9_TRAMO"
    ]

    # ConversiÃ³n a ENTERO REDONDEADO
    num_reg = int(round(len(df)))
    num_deudores = int(round(df["RUT"].nunique()))
    total_dia = int(round(df["MONTO_AL_DIA"].sum()))
    totales_mora = [int(round(df[c].sum())) for c in campos_mora]
    num_deud_mora = int(round(df[df[campos_mora].sum(axis=1) > 0]["RUT"].nunique()))
    total_lineas = int(round(df[df["TIPO_CREDITO"] == 42]["MONTO_ORIGINAL_OPERACION"].sum()))

    texto = f"""InstituciÃ³n (nombre): {institucion}   CÃ³digo: {codigo}
InformaciÃ³n correspondiente a la fecha ({fecha_ddmmaaaa}): Archivo RDC01.

NÃºmero registros informados (recuento simple): {num_reg}
NÃºmero de deudores vigentes (recuento de RUTs distintos): {num_deudores}
Total obligaciones reportables al dÃ­a y operaciones contingentes (suma campo 16): {total_dia}
Total obligaciones reportables con morosidad menor a 30 dÃ­as (suma campo 17): {totales_mora[0]}
Total obligaciones reportables con morosidad desde 30 a menos de 60 dÃ­as (suma campo 18): {totales_mora[1]}
Total obligaciones reportables con morosidad desde 60 a menos de 90 dÃ­as (suma campo 19): {totales_mora[2]}
Total obligaciones reportables con morosidad desde 90 dÃ­as a menos de 180 dÃ­as (suma campo 20): {totales_mora[3]}
Total obligaciones reportables con morosidad desde 180 dÃ­as a menos de un aÃ±o (suma campo 21): {totales_mora[4]}
Total obligaciones reportables con morosidad desde un aÃ±o a menos de dos aÃ±os (suma campo 22): {totales_mora[5]}
Total obligaciones reportables con morosidad desde dos aÃ±os a menos de tres aÃ±os (suma campo 23): {totales_mora[6]}
Total obligaciones reportables con morosidad desde tres aÃ±os a menos de cuatro aÃ±os (suma campo 24): {totales_mora[7]}
Total obligaciones reportables con morosidad superior a cuatro aÃ±os (suma campo 25): {totales_mora[8]}
NÃºmero de deudores con morosidad en cualquier tramo: {num_deud_mora}
Total lÃ­neas de crÃ©dito de libre disposiciÃ³n (suma campo 15 para campo 6 igual a 42): {total_lineas}
"""

    ruta_guardado.mkdir(parents=True, exist_ok=True)
    ruta_final = ruta_guardado / f"caratula_{fecha_ddmmaaaa.replace('-', '')}.txt"
    with open(ruta_final, "w", encoding="utf-8") as f:
        f.write(texto)

    return ruta_final
```
